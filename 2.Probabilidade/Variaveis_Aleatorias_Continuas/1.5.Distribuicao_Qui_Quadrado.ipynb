A distribuição qui-quadrado é um caso especial da distribuição gama, quando $\alpha = \frac{v}{2}$ e $\beta = 2$, onde _v_ é um número inteiro maior 0.(BUSSAB, MORETTIN, 2004)

Importar bibliotecas.

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('seaborn')
```

Função Gama:

$
\Gamma(\alpha) = \int_{0}^{\infty} e ^-x \cdot x^{\alpha - 1}dx,\ \alpha > 0.
$

```python
def gamma_function(alpha):
    expression = lambda x, aplha: np.exp(-x) * x ** (alpha - 1)
    return quad(expression, 0, np.inf, (alpha))
```

Função de Distribuição de Pobabilidade Qui-Quadrado:

$
f(x, v) = \frac{1}{\Gamma(\frac{v}{2}) \cdot 2^{\frac{v}{2}}} \cdot x^{\frac{v}{2} - 1} \cdot e^{\frac{-x}{2}},\ x > 0\ e\ y \in N.
$

```python
def chi_square_distribution(x, v):
    return ((gamma_function(v / 2)[0] * 2 ** (v / 2))) * (x ** ((v / 2) - 1)) * (np.exp(-x / 2))
```

Exemplos:

Dentro do intervalo do eixo x, onde $x > 0$, aplicamos seis valores diferentes para v para observarmos a diferenças no comportamento gráfico da distribuição qui-quadrado.

Na figura 1 podemos observar os diferentes _plots_.



```python
x_values = list(np.arange(0.01, 25, 0.01))
y_values_01 =list(map(lambda value: chi_square_distribution(value, 1), x_values))
y_values_02 =list(map(lambda value: chi_square_distribution(value, 2), x_values))
y_values_03 =list(map(lambda value: chi_square_distribution(value, 3), x_values))
y_values_04 =list(map(lambda value: chi_square_distribution(value, 4), x_values))
y_values_05 =list(map(lambda value: chi_square_distribution(value, 5), x_values))
y_values_06 =list(map(lambda value: chi_square_distribution(value, 6), x_values))
```

```python
figure = plt.figure(figsize=(20,15))
figure.suptitle('Figure 1. Chi square distribuiton examples.', size=20)
plots = figure.subplots(nrows=2, ncols=3)

plots[0][0].plot(x_values, y_values_01, label='Chi Square Distribution')
plots[0][0].set_title('Chi Square Distribuiton with v = 1.')
plots[0][0].text(20, 5, 'v = 1')
plots[0][0].set_xlabel('x values')
plots[0][0].set_ylabel('y values')
plots[0][0].legend()

plots[0][1].plot(x_values, y_values_02, label='Chi Square Distribution')
plots[0][1].set_title('Chi Square Distribuiton with v = 2.')
plots[0][1].text(20, 0.25, 'v = 2')
plots[0][1].set_xlabel('x values')
plots[0][1].set_ylabel('y values')
plots[0][1].legend()

plots[0][2].plot(x_values, y_values_03, label='Chi Square Distribution')
plots[0][2].set_title('Chi Square Distribuiton with v = 3.')
plots[0][2].text(20, 0.2, 'v = 3')
plots[0][2].set_xlabel('x values')
plots[0][2].set_ylabel('y values')
plots[0][2].legend()

plots[1][0].plot(x_values, y_values_04, label='Chi Square Distribution')
plots[1][0].set_title('Chi Square Distribuiton with v = 4.')
plots[1][0].text(20, 0.5, 'v = 4')
plots[1][0].set_xlabel('x values')
plots[1][0].set_ylabel('y values')
plots[1][0].legend()

plots[1][1].plot(x_values, y_values_05, label='Chi Square Distribution')
plots[1][1].set_title('Chi Square Distribuiton with v = 5.')
plots[1][1].text(20, 2, 'v = 5')
plots[1][1].set_xlabel('x values')
plots[1][1].set_ylabel('y values')
plots[1][1].legend()

plots[1][2].plot(x_values, y_values_06, label='Chi Square Distribution')
plots[1][2].set_title('Chi Square Distribuiton with v = 6.')
plots[1][2].text(20, 5, 'v = 6')
plots[1][2].set_xlabel('x values')
plots[1][2].set_ylabel('y values')
plots[1][2].legend()
```
