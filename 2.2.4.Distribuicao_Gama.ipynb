__Definição:__ "Em teoria das probabilidades e estatística, a distribuição gama é uma família de distribuições contínuas de probabilidade de dois parâmetros.[..] A distribuição gama é usada para modelar valores de dados positivos que são assimétricos à direita e maiores que 0. Ela é comumente usada em estudos de sobrevivência de confiabilidade."

Além disso, a Distribuição Gama é uma extensão do Modelo Exponencial, dado pelo caso específico m que alpha = 1. Outra situação importante é obtido através de alpha = v/2 e beta = 2, com v inteiro e maior que 0 que resulta na Distribuição Qui-Quadrado.

Uma outra observação interessante é que diferente do modelo normal a distribuição gama engloba as ditribuições que possuem assimetria à direita (como podemos observar na figura 1).

Importar bibliotecas.

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('seaborn')
```

Função Gama:

$
f(\alpha) = \int_{0}^{\infty} e ^-x \cdot x^{\alpha - 1}dx,\ \alpha > 0.
$

```python
def gamma_function(alpha):
    expression = lambda x, aplha: np.exp(-x) * x ** (alpha - 1)
    return quad(expression, 0, np.inf, (alpha))
```

Função de Distribuição de Pobabilidade Gamma:

$
f(x, \alpha, \beta) = \frac{1}{\Gamma(\alpha) \cdot \beta^\alpha} \cdot x^{\alpha - 1} \cdot e^{\frac{-x}{\beta}},\ x > 0;\\
f(x, \alpha, \beta) = 0,\ x < 0.
$

```python
def gamma_distribution(x, alpha, beta):
    numerator = x ** (alpha - 1) * np.exp(-x / beta)
    denominator = gamma_function(alpha)[0] * beta ** alpha
    return numerator / denominator
```

Exemplo:

Dado $\alpha = 3$ e $\beta = 1$, obtemos o gráfico da figura 1.

```python
alpha = 3
beta = 1
```

```python
x_values = list(np.arange(0, 10, 0.01))
y_values =list(map(lambda value: gamma_distribution(value, alpha, beta), x_values))
```

```python
max_value_index = y_values.index(max(y_values))
```

```python
plt.plot(x_values, y_values, label='Gamma Distribution')
plt.title('1.Gamma Distribuiton')
plt.vlines(x_values[max_value_index], 0, y_values[max_value_index], ls='dashed', color='black')
plt.xlabel('x values')
plt.ylabel('y values')
plt.text(9, 0.06, f'alpha = {alpha}')
plt.text(9, 0.05, f'beta = {beta}')
plt.legend()
```
